```python
import pandas as pd

path = "../dados/TMDB_all_movies.csv"

df_movies = pd.read_csv(path)
df_movies.columns
```

```python
df_movies["overview"].head()
```

```python
df_movies["tagline"].head()
```

```python
def remove_virgula(texto):
    return texto.replace(',', '') if isinstance(texto, str) else texto

# generos unicos
generos = set()
for lista in df_movies["genres"].dropna().str.split():
    for genero in lista:
        if genero:  # evita strings vazias

            generos.add(remove_virgula(genero))
lista_generos = sorted(generos)

lista_generos

```
