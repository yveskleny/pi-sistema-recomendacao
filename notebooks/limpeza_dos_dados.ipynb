```python
import pandas as pd
import os

caminho_relativo_dados = "../dados/TMDB_all_movies.csv"

caminho_absoluto_dados = os.path.abspath(caminho_relativo_dados)

filmes_df = pd.read_csv(caminho_absoluto_dados)
```

```python

# colunas_desejadas = [
#     "id",
#     "title",
#     "vote_average",
#     "vote_count",
#     "revenue",
#     "budget",
#     "original_language",
#     "popularity",
#     "genres",
#     "production_companies",
#     "cast",
#     "director",
#     "director_of_photography",
#     "writers",
#     "producers",
#     "music_composer",
#     "imdb_rating",
#     "imdb_votes"
# ]
colunas_desejadas = [
    "id",
    "title",
    "vote_average",
    "vote_count",
    "revenue",
    "budget",
    "original_language",
    "genres",
    "cast",
    "director",
    "imdb_rating",
    "imdb_votes"
]
movies_df = filmes_df[colunas_desejadas].dropna(subset=[
    "id",
    "title",
    "vote_average",
    "vote_count",
    "revenue",
    "budget",
    "original_language",
    "genres",
    "cast",
    "director",
    "imdb_rating",
    "imdb_votes"
])
movies_df.info()
```

```python
movies_df.info()
```

```python
movies_df = movies_df[(movies_df["vote_count"] > 100) | (movies_df["imdb_votes"] > 100)]
```

```python
pasta_dados = os.path.abspath("../dados")
path = os.path.join(pasta_dados, "dados_processados.csv")
movies_df.to_csv(path)
```
